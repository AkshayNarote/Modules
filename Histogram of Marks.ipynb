```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("D:\\Data Science by SRK\\studentData.xlsx")
print(df)

n,bins,patches = plt.hist(df['Marks'],bins=10,range=(0,100))   #It will give 10 bins of size 100/10 
print(n,bins,patches)   # It will give array of n (n = no. of students in perticular range(frequency)), array of bins, patches = BarContainer object of 10 artists  
plt.bar_label(patches)  # It will put numbers on top of bars as frequency count in that perticular bars
plt.xlabel("Marks")     # It gives label name to x-axis
plt.ylabel("No. of Students/Frequency")     # It gives label name to y-axis
plt.title("Histogram of Marks")    # It gives title name to graph
plt.xticks(list(range(0,101,10)))  # It puts values of bars number on x-axis(ex:0,10,20,30,...)
plt.show()              # It will show only graph of histogram in output, not other thinks related to it
```
